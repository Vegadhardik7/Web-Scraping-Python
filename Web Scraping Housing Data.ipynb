```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.pararius.com/apartments/amsterdam?ac=1"

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
lists = soup.find_all('section', class_ = "listing-search-item")

for i in lists:
    title = i.find('a', class_ = "listing-search-item__link--title").get_text().replace('\n','').lstrip().rstrip()
    location = i.find('div', class_ = "listing-search-item__location").get_text().replace('\n','').lstrip().rstrip()
    price = i.find('div', class_ = "listing-search-item__price").get_text().replace('\n','').lstrip().rstrip()
    area = i.find('li', class_ = "illustrated-features__item--surface-area").get_text().replace('\n','').lstrip().rstrip()
    
    lst = [title, location, price, area]
    print(lst)
```

```python
from csv import writer
with open('housing.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title', 'Location', 'Price', 'Area']
    thewriter.writerow(header)
                       
    for i in lists:
        title = i.find('a', class_ = "listing-search-item__link--title").get_text().replace('\n','').lstrip().rstrip()
        location = i.find('div', class_ = "listing-search-item__location").get_text().replace('\n','').lstrip().rstrip()
        price = i.find('div', class_ = "listing-search-item__price").get_text().replace('\n','').lstrip().rstrip()
        area = i.find('li', class_ = "illustrated-features__item--surface-area").get_text().replace('\n','').lstrip().rstrip()

        
        info = [title, location, price, area]
        thewriter.writerow(info)
```
