```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=clothing&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"

page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
print(soup.title.get_text())
```

```python
lists = soup.find_all('div', class_ = "_13oc-S") # _13oc-S

for i in lists:
    title = i.find('div', class_ = "_2WkVRV").get_text().replace('\n','').lstrip().rstrip()
    price = i.find('div', class_ = "_30jeq3").get_text().replace('\n','').lstrip().rstrip()
    
    
    lst = [title, price]
    print(lst)
```
